```python
# We got some issue here in above dest_region column, let's correct it.

airlines['dest_region'] = airlines['dest_region'].str.replace('eur', 'Europe')
airlines['dest_region'] = airlines['dest_region'].str.replace('EAST US', 'East US')
airlines['dest_region'] = airlines['dest_region'].str.replace('middle east', 'Middle East')
```

```python
#Now checking for confirmation.
print(airlines.dest_region.unique())
print()

print(airlines['dest_region'].value_counts(sort=True))
```

```python
#Targetting dest_size column

print(airlines.dest_size.unique())
print()
print()
print(airlines['dest_size'].value_counts(sort=True))
print()
```

```python
# We got some issue here in above dest_size column, let's correct it.
airlines['dest_size'] = airlines['dest_size'].str.strip()
```

```python
print(airlines.dest_size.unique())
print()
print()
print(airlines['dest_size'].value_counts(sort=True))
print()
```

```python
airlines.head()
```

```python
airlines['wait_min'].value_counts(sort=True)

#here we can see, it's difficult to see how long someone has to wait from categorical point of view, we can change it to
# categories so that passenger can easily get the idea.
```

```python
# Create ranges for categories
label_ranges =[0,60,100,500, np.inf]
label_names = ['short','medium','long', 'too long']

airlines['wait_category'] = pd.cut(airlines['wait_min'], bins = label_ranges, labels = label_names)
airlines.head()
```

```python
airlines.day.unique()
```

```python
#similary with weekdays but with a different method (mapping)
# Create mappings and replace
mapping = { 'Monday': 'weekday',
            'Tuesday' : 'weekday',
            'Wednesday': 'weekday',
            'Thursday': 'weekday',
            'Friday': 'weekday',
            'Saturday':'Weekend',
            'Sunday': 'Weekend'
          }

airlines['Weekday_Weekend'] = airlines['day'].replace(mapping)
airlines.head()
```

```python
airlines.Weekday_Weekend.unique()
```

```python
# Store length of each row in survey_response column

# here we are storing the length of value of wait_category in each row
resp_length =airlines['wait_category'].str.len()
# print(resp_length)

# it is fetching the whole dataframe but only where resp_length is greater than 5.
airline_survey = airlines[resp_length > 5]
airline_survey.head()
# airline_survey[['wait_min','wait_category']]
```
